# CÂU 1 (2 ĐIỂM) – Python Cơ Bản

Viết chương trình Python thực hiện **quản lý danh sách sinh viên** với các yêu cầu sau:

**a) (0.5đ)** Định nghĩa class `SinhVien` có các thuộc tính: `mssv` (str), `ho_ten` (str), `nam_sinh` (int), `diem_tb` (float). Class phải có phương thức `__str__` trả về thông tin sinh viên theo định dạng:
```
[MSSV] Họ tên | Năm sinh: xxxx | Điểm TB: x.x
```

**b) (0.5đ)** Tạo danh sách ít nhất **5 sinh viên** (dữ liệu tự chọn). Sử dụng list comprehension để lọc ra danh sách sinh viên có **điểm trung bình >= 5.0** (đủ điều kiện lên lớp).

**c) (0.5đ)** Viết hàm `thong_ke(ds_sv)` nhận vào danh sách sinh viên, trả về dictionary chứa:
- `'tong_sv'`: tổng số sinh viên
- `'sv_len_lop'`: số sinh viên đủ điều kiện lên lớp (điểm TB >= 5.0)
- `'diem_cao_nhat'`: điểm trung bình cao nhất
- `'diem_trung_binh_lop'`: điểm trung bình của cả lớp

**d) (0.5đ)** In ra kết quả thống kê theo định dạng rõ ràng, đồng thời in danh sách sinh viên **từ 20 tuổi trở lên** (tính theo năm hiện tại là 2025).

````python
 #Cau 1
#a)Định nghĩa class SinhVien
class SinhVien:
    def __init__(self, mssv, ho_ten, nam_sinh, diem_tb):
        self.mssv = mssv
        self.ho_ten = ho_ten
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb

    def __str__(self):
        return f"[{self.mssv}] {self.ho_ten} | Năm sinh: {self.nam_sinh} | Điểm TB: {self.diem_tb}"

#b) Tạo danh sách sinh viên + lọc sinh viên đủ điều kiện lên lớp
# Danh sách sinh viên
ds_sv = [
    SinhVien("SV001", "Nguyễn Văn A", 2003, 6.5),
    SinhVien("SV002", "Trần Thị B", 2006, 4.8),
    SinhVien("SV003", "Lê Văn C", 2004, 7.2),
    SinhVien("SV004", "Phạm Thị D", 2005, 5.0),
    SinhVien("SV005", "Hoàng Văn E", 2002, 3.9)
]

# Lọc sinh viên có điểm TB >= 5.0
sv_len_lop = [sv for sv in ds_sv if sv.diem_tb >= 5.0]

#c) Hàm thống kê
def thong_ke(ds_sv):
    tong_sv = len(ds_sv)
    sv_len_lop = len([sv for sv in ds_sv if sv.diem_tb >= 5.0])
    diem_cao_nhat = max(sv.diem_tb for sv in ds_sv)
    diem_trung_binh_lop = sum(sv.diem_tb for sv in ds_sv) / tong_sv

    return {
        'tong_sv': tong_sv,
        'sv_len_lop': sv_len_lop,
        'diem_cao_nhat': diem_cao_nhat,
        'diem_trung_binh_lop': diem_trung_binh_lop
    }

    #d) In kết quả thống kê + sinh viên ≥ 20 tuổi (năm 2025)
    # Gọi hàm thống kê
tk = thong_ke(ds_sv)

# In kết quả
print("=== THỐNG KÊ LỚP ===")
print(f"Tổng số sinh viên: {tk['tong_sv']}")
print(f"Số SV lên lớp: {tk['sv_len_lop']}")
print(f"Điểm cao nhất: {tk['diem_cao_nhat']}")
print(f"Điểm trung bình lớp: {tk['diem_trung_binh_lop']:.2f}")

# Sinh viên từ 20 tuổi trở lên (2025)
print("\n=== SINH VIÊN ≥ 20 TUỔI ===")
for sv in ds_sv:
    tuoi = 2025 - sv.nam_sinh
    if tuoi >= 20:
        print(sv)
````

# CÂU 2 (2.5 ĐIỂM) – Thuật Toán Tìm Kiếm A*

Cho đồ thị vô hướng có trọng số như sau (chi phí di chuyển giữa các đỉnh):

```
        A
       / \
      2   4
     /     \
    B       C
   / \     / \
  3   1   2   5
 /     \ /     \
D       E       F
         \
          3
           \
            G  (Goal)
```

**Danh sách cạnh (cạnh hai chiều):**
```
A-B: 2,  A-C: 4
B-D: 3,  B-E: 1
C-E: 2,  C-F: 5
E-G: 3
```

**Giá trị heuristic h(n) (ước lượng khoảng cách đến đích G):**
```
h(A)=7, h(B)=5, h(C)=6, h(D)=9, h(E)=3, h(F)=8, h(G)=0
```

**a) (1.0đ)** Cài đặt thuật toán A* để tìm đường đi ngắn nhất từ đỉnh **A** đến đỉnh **G**. Sử dụng danh sách `OPEN` và `CLOSE`. Với mỗi bước, in ra trạng thái của `OPEN` (node, f, g, h).

**b) (1.0đ)** In ra đường đi tối ưu tìm được và tổng chi phí thực tế.

**c) (0.5đ)** So sánh thuật toán A* với BFS (tìm kiếm theo chiều rộng): A* có ưu điểm gì? Viết câu trả lời dưới dạng comment trong cell bên dưới.

````python
#Câu 2
#a) Cài đặt thuật toán A* + in trạng thái OPEN
import heapq

# Đồ thị (undirected graph)
graph = {
    'A': {'B': 2, 'C': 4},
    'B': {'A': 2, 'D': 3, 'E': 1},
    'C': {'A': 4, 'E': 2, 'F': 5},
    'D': {'B': 3},
    'E': {'B': 1, 'C': 2, 'G': 3},
    'F': {'C': 5},
    'G': {'E': 3}
}

# Heuristic
h = {
    'A': 7, 'B': 5, 'C': 6,
    'D': 9, 'E': 3, 'F': 8, 'G': 0
}

def astar(start, goal):
    OPEN = []  # priority queue (min-heap)
    heapq.heappush(OPEN, (h[start], start))  # (f, node)

    g = {start: 0}
    parent = {start: None}
    CLOSE = set()

    while OPEN:
        # In trạng thái OPEN
        print("\nOPEN:")
        # Sort OPEN for consistent output during printing, although heapq doesn't keep it sorted
        # We need to make a copy to sort, not modify the original heap
        open_copy = sorted(OPEN, key=lambda x: x[0])
        for f_val, node in open_copy:
            # Ensure node is in g before trying to access it (can happen with initial push before g[start] is set, but here it's fine)
            g_val = g.get(node, 'N/A')
            print(f"{node}: f={f_val}, g={g_val}, h={h[node]}")

        f_current, current = heapq.heappop(OPEN)

        if current == goal:
            # reconstruct path
            path = []
            while current:
                path.append(current)
                current = parent[current]
            return path[::-1], g[goal]

        CLOSE.add(current)

        for neighbor, cost in graph[current].items():
            if neighbor in CLOSE:
                continue

            tentative_g = g[current] + cost

            if neighbor not in g or tentative_g < g[neighbor]:
                g[neighbor] = tentative_g
                f = tentative_g + h[neighbor]
                parent[neighbor] = current
                heapq.heappush(OPEN, (f, neighbor))

    return None, float('inf')

#b) In đường đi tối ưu và chi phí
path, cost = astar('A', 'G')

print("\n=== KẾT QUẢ ===")
if path:
    print("Đường đi tối ưu:", " -> ".join(path))
    print("Tổng chi phí:", cost)
else:
    print("Không tìm thấy đường đi.")

#c) So sánh A* và BFS (viết dưới dạng comment)
# So sánh A* và BFS:
# - BFS không xét trọng số và không dùng heuristic → chỉ đảm bảo tối ưu khi các cạnh có cùng chi phí.
# - A* sử dụng hàm f(n) = g(n) + h(n):
#     + g(n): chi phí thực tế từ start đến n
#     + h(n): ước lượng từ n đến goal
# - Nhờ heuristic, A* ưu tiên các node "có vẻ gần đích hơn"
# → Tìm đường đi tối ưu nhanh hơn, mở rộng ít node hơn so với BFS.
# - Nếu heuristic tốt (admissible), A* vẫn đảm bảo tối ưu.
````

# CÂU 3 (2.5 ĐIỂM) – Thuật Toán Minimax cho TicTacToe

Cài đặt trò chơi **TicTacToe 3x3** với AI sử dụng thuật toán **Minimax**.

**a) (0.5đ)** Cài đặt các hàm khởi tạo và kiểm tra trạng thái:
- `initial_state()`: trả về bảng 3x3 khởi tạo với giá trị `None`
- `player(board)`: xác định lượt đi (`'X'` hoặc `'O'`)
- `actions(board)`: trả về tập hợp các nước đi hợp lệ `(i, j)`
- `result(board, action)`: trả về bảng mới sau khi thực hiện nước đi (dùng `copy.deepcopy`)

**b) (0.5đ)** Cài đặt hàm đánh giá kết quả:
- `winner(board)`: trả về `'X'`, `'O'` hoặc `None`
- `terminal(board)`: trả về `True` nếu trò chơi kết thúc
- `utility(board)`: trả về `1` (X thắng), `-1` (O thắng), `0` (hòa)

**c) (1.0đ)** Cài đặt thuật toán Minimax:
- `max_value(state)`: tính giá trị MAX (người chơi X – muốn tối đa hóa)
- `min_value(state)`: tính giá trị MIN (AI O – muốn tối thiểu hóa)
- `minimax(board)`: trả về nước đi tối ưu cho lượt hiện tại

**d) (0.5đ)** Viết vòng lặp trò chơi đơn giản:
- Người chơi nhập tọa độ `(hàng, cột)` theo cú pháp: `1 1` (tách bởi dấu cách)
- AI tự động tính nước đi tối ưu bằng Minimax
- In bảng sau mỗi lượt; thông báo kết quả khi trò chơi kết thúc

````python
#a) Khởi tạo & hàm trạng thái
import copy

def initial_state():
    return [[None, None, None],
            [None, None, None],
            [None, None, None]]

def player(board):
    # X đi trước
    flat = [cell for row in board for cell in row]
    if flat.count('X') <= flat.count('O'):
        return 'X'
    return 'O'

def actions(board):
    return {(i, j) for i in range(3) for j in range(3) if board[i][j] is None}

def result(board, action):
    i, j = action
    if board[i][j] is not None:
        raise Exception("Nước đi không hợp lệ!")
    new_board = copy.deepcopy(board)
    new_board[i][j] = player(board)
    return new_board

#b) Kiểm tra kết quả
def winner(board):
    lines = []

    # Hàng
    lines.extend(board)
    # Cột
    for j in range(3):
        lines.append([board[i][j] for i in range(3)])
    # Chéo
    lines.append([board[i][i] for i in range(3)])
    lines.append([board[i][2 - i] for i in range(3)])

    for line in lines:
        if line[0] is not None and line.count(line[0]) == 3:
            return line[0]
    return None

def terminal(board):
    if winner(board) is not None:
        return True
    return all(cell is not None for row in board for cell in row)

def utility(board):
    w = winner(board)
    if w == 'X':
        return 1
    elif w == 'O':
        return -1
    return 0

#c) Thuật toán Minimax
def max_value(board):
    if terminal(board):
        return utility(board)
    v = float('-inf')
    for action in actions(board):
        v = max(v, min_value(result(board, action)))
    return v

def min_value(board):
    if terminal(board):
        return utility(board)
    v = float('inf')
    for action in actions(board):
        v = min(v, max_value(result(board, action)))
    return v

def minimax(board):
    if terminal(board):
        return None

    current = player(board)

    if current == 'X':
        best_val = float('-inf')
        best_move = None
        for action in actions(board):
            val = min_value(result(board, action))
            if val > best_val:
                best_val = val
                best_move = action
        return best_move
    else:
        best_val = float('inf')
        best_move = None
        for action in actions(board):
            val = max_value(result(board, action))
            if val < best_val:
                best_val = val
                best_move = action
        return best_move

#d) Vòng lặp trò chơi (CLI đơn giản)
def print_board(board):
    for row in board:
        print([ " " if cell is None else cell for cell in row])
    print()

def play():
    board = initial_state()

    while not terminal(board):
        print_board(board)

        if player(board) == 'X':
            # Người chơi
            try:
                i, j = map(int, input("Nhập nước đi (hàng cột, ví dụ 1 1): ").split())
                i -= 1
                j -= 1
                if (i, j) not in actions(board):
                    print("Nước đi không hợp lệ, thử lại!")
                    continue
                board = result(board, (i, j))
            except:
                print("Input sai định dạng!")
                continue
        else:
            # AI (O) sử dụng Alpha-Beta Pruning
            print("AI đang suy nghĩ (Alpha-Beta)...")
            # find_best_move được thiết kế để tìm nước đi tốt nhất cho 'O' (MIN)
            move = find_best_move(board)
            board = result(board, move)

    print_board(board)
    w = winner(board)
    if w:
        print(f"Người thắng: {w}")
    else:
        print("Hòa!")

# Chạy game
if __name__ == "__main__":
    play()
````

# CÂU 4 (3 ĐIỂM) – Alpha-Beta Pruning

Cải tiến chương trình TicTacToe ở Câu 3 bằng cách thay thế Minimax thuần túy bằng **Minimax kết hợp Alpha-Beta Pruning**.

**a) (1.0đ)** Cài đặt hàm `minimax_ab(board, depth, alpha, beta, is_maximizing)` với các tham số:
- `board`: trạng thái bảng hiện tại
- `depth`: độ sâu hiện tại (dùng để ưu tiên thắng nhanh / thua muộn)
- `alpha`: cận dưới của nút MAX (`-math.inf` ban đầu)
- `beta`: cận trên của nút MIN (`+math.inf` ban đầu)
- `is_maximizing`: `True` nếu đang tính lượt MAX

Yêu cầu:
- Điểm trả về: `10 - depth` (nút MAX thắng), `-10 + depth` (nút MIN thắng), `0` (hòa)
- Cắt tỉa khi `beta <= alpha`

**b) (0.5đ)** Cài đặt hàm `find_best_move(board)` trả về nước đi tốt nhất cho AI (người chơi O – MIN) bằng cách gọi `minimax_ab`.

**c) (1.0đ)** So sánh hiệu năng giữa Minimax và Alpha-Beta Pruning:
- Thêm bộ đếm `node_count` vào cả hai hàm để đếm số lần gọi đệ quy
- Chạy cả hai từ trạng thái khởi đầu và so sánh số node được duyệt
- In kết quả so sánh

**d) (0.5đ)** Tích hợp `find_best_move` vào vòng lặp trò chơi (thay thế `minimax` ở Câu 3) và chạy thử. Nhận xét bằng comment: Alpha-Beta Pruning có tác động thế nào đến tốc độ phản hồi của AI?

````python
#a) Cài đặt minimax_ab
import math
import copy

node_count_ab = 0  # đếm số node duyệt (Alpha-Beta)

def minimax_ab(board, depth, alpha, beta, is_maximizing):
    global node_count_ab
    node_count_ab += 1

    w = winner(board)
    if w == 'X':
        return 10 - depth
    elif w == 'O':
        return -10 + depth
    elif terminal(board):
        return 0

    if is_maximizing:  # X
        max_eval = -math.inf
        for action in actions(board):
            eval = minimax_ab(result(board, action), depth + 1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:  # cắt tỉa
                break
        return max_eval
    else:  # O
        min_eval = math.inf
        for action in actions(board):
            eval = minimax_ab(result(board, action), depth + 1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:  # cắt tỉa
                break
        return min_eval

#b) Hàm tìm nước đi tốt nhất cho AI (O – MIN)
def find_best_move(board):
    best_val = math.inf
    best_move = None

    for action in actions(board):
        # AI (O) là người chơi MIN, muốn giảm giá trị, nên khi gọi minimax_ab, is_maximizing = True để nó tính giá trị cho người chơi X (MAX)
        # Rồi sau đó mình sẽ chọn nước đi của AI (O) có move_val nhỏ nhất
        move_val = minimax_ab(result(board, action), 0, -math.inf, math.inf, True) # is_maximizing=True vì đây là lượt của X sau khi O đã đánh
        if move_val < best_val:
            best_val = move_val
            best_move = action

    return best_move

#c) So sánh Minimax vs Alpha-Beta
node_count_mm = 0  # đếm node Minimax thường

def minimax_count(board, is_maximizing):
    global node_count_mm
    node_count_mm += 1

    if terminal(board):
        return utility(board)

    if is_maximizing:
        v = -math.inf
        for action in actions(board):
            v = max(v, minimax_count(result(board, action), False))
        return v
    else:
        v = math.inf
        for action in actions(board):
            v = min(v, minimax_count(result(board, action), True))
        return v


# So sánh từ trạng thái ban đầu
board0 = initial_state()

# Minimax thường
node_count_mm = 0
minimax_count(board0, True)

# Alpha-Beta
node_count_ab = 0
minimax_ab(board0, 0, -math.inf, math.inf, True)

print("=== SO SÁNH ===")
print(f"Số node Minimax thường: {node_count_mm}")
print(f"Số node Alpha-Beta: {node_count_ab}")



#d) Tích hợp vào game
# Thay đoạn AI:
#else:
#    print("AI đang suy nghĩ (Alpha-Beta)...")
#    move = find_best_move(board)
#    board = result(board, move)


# Nhận xét:
# Alpha-Beta Pruning giúp giảm đáng kể số node cần duyệt so với Minimax thuần túy
# bằng cách cắt bỏ các nhánh không cần thiết (beta <= alpha).
# → AI phản hồi nhanh hơn rõ rệt, đặc biệt khi cây tìm kiếm lớn.
# Tuy nhiên, kết quả (nước đi tối ưu) vẫn giữ nguyên như Minimax.
````
