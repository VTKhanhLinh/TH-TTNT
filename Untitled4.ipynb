```python
class Graph:
    def __init__(self, adjac_lis, heuristic_dict):
        self.adjac_lis = adjac_lis
        self.h_val = heuristic_dict

    def get_neighbors(self, v):
        return self.adjac_lis.get(v, [])

    def h(self, n):
        return self.h_val.get(n, 0)

    def a_star_algorithm(self, start, stop):
        open_lst = set([start])
        closed_lst = set([])

        g = {start: 0} # Chi phí thực tế từ xuất phát
        par = {start: start}

        while open_lst:
            n = None
            # Chọn nút có f(n) = g(n) + h(n) nhỏ nhất
            for v in open_lst:
                if n == None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v

            if n == stop:
                path = []
                while par[n] != n:
                    path.append(n)
                    n = par[n]
                path.append(start)
                path.reverse()
                print(f'Đường đi tìm thấy: {path}')
                return path

            if n == None:
                print('Không tìm thấy đường đi!')
                return None

            for (m, weight) in self.get_neighbors(n):
                if m not in open_lst and m not in closed_lst:
                    open_lst.add(m)
                    par[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        par[m] = n
                        if m in closed_lst:
                            closed_lst.remove(m)
                            open_lst.add(m)

            open_lst.remove(n)
            closed_lst.add(n)

        print('Không có đường đi!')
        return None

# Khai báo đồ thị
adjac_lis = {
    'A': [('B', 5), ('C', 10)],
    'B': [('D', 2), ('E', 1)],
    'C': [('E', 5)],
    'D': [('F', 3)],
    'E': [('F', 8)],
    'F': []
}

# Heuristic (Khoảng cách ước lượng đến F)
h_dist = {'A': 10, 'B': 6, 'C': 4, 'D': 3, 'E': 5, 'F': 0}

graph = Graph(adjac_lis, h_dist)
graph.a_star_algorithm('A', 'F')
```
