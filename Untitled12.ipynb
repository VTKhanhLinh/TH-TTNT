```python
#Bài 4. Cài đặt thuật toán cắt tỉa Alpha-beta cho bài toán tictactoe với ma trận 5x5, 10x10, NxN.
import math

N = 5          # đổi thành 10 hoặc N bất kỳ
WIN_LEN = 4    # số quân liên tiếp để thắng (có thể chỉnh)
MAX_DEPTH = 2  # độ sâu (tăng = mạnh hơn nhưng chậm hơn)

board = [[" "]*N for _ in range(N)]

# =====================
# KIỂM TRA THẮNG
# =====================
def check_winner(b):
    # hàng
    for i in range(N):
        for j in range(N - WIN_LEN + 1):
            line = b[i][j:j+WIN_LEN]
            if line[0] != " " and all(x == line[0] for x in line):
                return line[0]

    # cột
    for j in range(N):
        for i in range(N - WIN_LEN + 1):
            line = [b[i+k][j] for k in range(WIN_LEN)]
            if line[0] != " " and all(x == line[0] for x in line):
                return line[0]

    # chéo \
    for i in range(N - WIN_LEN + 1):
        for j in range(N - WIN_LEN + 1):
            line = [b[i+k][j+k] for k in range(WIN_LEN)]
            if line[0] != " " and all(x == line[0] for x in line):
                return line[0]

    # chéo /
    for i in range(WIN_LEN-1, N):
        for j in range(N - WIN_LEN + 1):
            line = [b[i-k][j+k] for k in range(WIN_LEN)]
            if line[0] != " " and all(x == line[0] for x in line):
                return line[0]

    return None

def is_full(b):
    return all(cell != " " for row in b for cell in row)

# =====================
# HEURISTIC (đơn giản)
# =====================
def evaluate(b):
    winner = check_winner(b)
    if winner == "X":
        return 1000
    elif winner == "O":
        return -1000
    return 0

# =====================
# SINH NƯỚC ĐI (tối ưu)
# chỉ xét ô gần quân đã đánh
# =====================
def get_moves(b):
    moves = set()

    for i in range(N):
        for j in range(N):
            if b[i][j] != " ":
                for dx in [-1, 0, 1]:
                    for dy in [-1, 0, 1]:
                        ni, nj = i + dx, j + dy
                        if 0 <= ni < N and 0 <= nj < N and b[ni][nj] == " ":
                            moves.add((ni, nj))

    # nếu bàn trống → đánh giữa
    if not moves:
        return [(N//2, N//2)]

    return list(moves)

# =====================
# 🔥 ALPHA-BETA
# =====================
def minimax_ab(b, depth, alpha, beta, is_max):
    score = evaluate(b)

    if abs(score) == 1000 or depth == 0 or is_full(b):
        return score

    moves = get_moves(b)

    if is_max:
        best = -math.inf
        for i, j in moves:
            b[i][j] = "X"
            val = minimax_ab(b, depth-1, alpha, beta, False)
            b[i][j] = " "

            best = max(best, val)
            alpha = max(alpha, best)

            if beta <= alpha:
                break   # 🔥 cắt tỉa

        return best

    else:
        best = math.inf
        for i, j in moves:
            b[i][j] = "O"
            val = minimax_ab(b, depth-1, alpha, beta, True)
            b[i][j] = " "

            best = min(best, val)
            beta = min(beta, best)

            if beta <= alpha:
                break   # 🔥 cắt tỉa

        return best

# =====================
# TÌM NƯỚC ĐI TỐT NHẤT
# =====================
def best_move(b):
    best_val = -math.inf
    move = (-1, -1)

    for i, j in get_moves(b):
        b[i][j] = "X"
        move_val = minimax_ab(b, MAX_DEPTH-1, -math.inf, math.inf, False)
        b[i][j] = " "

        if move_val > best_val:
            best_val = move_val
            move = (i, j)

    return move
```
