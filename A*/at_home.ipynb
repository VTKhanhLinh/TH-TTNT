```python
import heapq

class TSP_AStar:
    def __init__(self, matrix):
        self.matrix = matrix
        self.n = len(matrix)

    def heuristic(self, curr_node, unvisited):
        """
        Hàm h(n): Ước lượng chi phí từ đỉnh hiện tại đi qua hết
        các đỉnh chưa thăm và quay về đỉnh xuất phát.
        Ở đây dùng: Cạnh nhỏ nhất còn lại.
        """
        if not unvisited:
            return self.matrix[curr_node][0] # Quay về đỉnh xuất phát (0)

        # Ước lượng đơn giản: lấy cạnh nhỏ nhất trong số các đỉnh chưa thăm
        min_edge = min([self.matrix[curr_node][v] for v in unvisited])
        return min_edge

    def solve(self, start_node=0):
        # Priority Queue lưu: (f, g, current_node, visited_mask, path)
        # f = g + h
        # visited_mask: dùng bitmask để lưu các đỉnh đã đi qua
        initial_unvisited = set(range(1, self.n))
        h = self.heuristic(start_node, initial_unvisited)

        pq = [(h, 0, start_node, 1 << start_node, [start_node])]

        # Lưu trữ các trạng thái đã thăm để tránh tính toán lại (u, mask)
        visited_states = {}

        while pq:
            f, g, u, mask, path = heapq.heappop(pq)

            # Nếu đã đi qua tất cả các đỉnh
            if mask == (1 << self.n) - 1:
                final_cost = g + self.matrix[u][start_node]
                path.append(start_node)
                print(f"--- ĐÃ TÌM THẤY LỜI GIẢI TSP ---")
                print(f"Lộ trình: {' -> '.join(map(str, path))}")
                print(f"Tổng chi phí: {final_cost}")
                return path

            # Tránh duyệt lại trạng thái tệ hơn
            state = (u, mask)
            if state in visited_states and visited_states[state] <= g:
                continue
            visited_states[state] = g

            # Thử đi đến các đỉnh chưa thăm
            for v in range(self.n):
                if not (mask & (1 << v)): # Nếu v chưa được thăm
                    new_g = g + self.matrix[u][v]
                    new_mask = mask | (1 << v)

                    unvisited = [i for i in range(self.n) if not (new_mask & (1 << i))]
                    new_h = self.heuristic(v, unvisited)

                    heapq.heappush(pq, (new_g + new_h, new_g, v, new_mask, path + [v]))

        print("Không tìm thấy đường đi!")
        return None

# --- THIẾT LẬP DỮ LIỆU ---
# Ma trận chi phí giữa các thành phố (0, 1, 2, 3)
# matrix[i][j] là chi phí đi từ i đến j
dist_matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

tsp = TSP_AStar(dist_matrix)
tsp.solve(0)
```
