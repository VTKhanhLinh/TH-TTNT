```python
import math

board = [[" "]*3 for _ in range(3)]

def check_winner(b):
    for row in b:
        if row[0] == row[1] == row[2] and row[0] != " ":
            return row[0]

    for col in range(3):
        if b[0][col] == b[1][col] == b[2][col] and b[0][col] != " ":
            return b[0][col]

    if b[0][0] == b[1][1] == b[2][2] and b[0][0] != " ":
        return b[0][0]
    if b[0][2] == b[1][1] == b[2][0] and b[0][2] != " ":
        return b[0][2]

    return None

def is_full(b):
    return all(cell != " " for row in b for cell in row)

# 🔥 Alpha-Beta Pruning
def minimax_ab(b, depth, alpha, beta, is_max):
    winner = check_winner(b)

    if winner == "X":
        return 1
    elif winner == "O":
        return -1
    elif is_full(b):
        return 0

    if is_max:  # AI (X)
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == " ":
                    b[i][j] = "X"
                    val = minimax_ab(b, depth+1, alpha, beta, False)
                    b[i][j] = " "

                    best = max(best, val)
                    alpha = max(alpha, best)

                    # 🔥 Cắt tỉa
                    if beta <= alpha:
                        return best
        return best

    else:  # Người chơi (O)
        best = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == " ":
                    b[i][j] = "O"
                    val = minimax_ab(b, depth+1, alpha, beta, True)
                    b[i][j] = " "

                    best = min(best, val)
                    beta = min(beta, best)

                    # 🔥 Cắt tỉa
                    if beta <= alpha:
                        return best
        return best

# Tìm nước đi tốt nhất
def best_move(b):
    best_val = -math.inf
    move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if b[i][j] == " ":
                b[i][j] = "X"
                move_val = minimax_ab(b, 0, -math.inf, math.inf, False)
                b[i][j] = " "

                if move_val > best_val:
                    best_val = move_val
                    move = (i, j)

    return move

# Test
board[0][0] = "X"
board[1][1] = "O"

print("Best move:", best_move(board))
```

```python
import math

board = [[" "]*3 for _ in range(3)]

def print_board():
    print("\n")
    for i, row in enumerate(board):
        print(" | ".join(row))
        if i < 2:
            print("--+---+--")
    print("\n")

def check_winner(b):
    for row in b:
        if row[0] == row[1] == row[2] and row[0] != " ":
            return row[0]

    for col in range(3):
        if b[0][col] == b[1][col] == b[2][col] and b[0][col] != " ":
            return b[0][col]

    if b[0][0] == b[1][1] == b[2][2] and b[0][0] != " ":
        return b[0][0]
    if b[0][2] == b[1][1] == b[2][0] and b[0][2] != " ":
        return b[0][2]

    return None

def is_full(b):
    return all(cell != " " for row in b for cell in row)

# 🔥 Alpha-Beta
def minimax_ab(b, alpha, beta, is_max):
    winner = check_winner(b)

    if winner == "X":
        return 1
    if winner == "O":
        return -1
    if is_full(b):
        return 0

    if is_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == " ":
                    b[i][j] = "X"
                    best = max(best, minimax_ab(b, alpha, beta, False))
                    b[i][j] = " "
                    alpha = max(alpha, best)

                    if beta <= alpha:
                        break
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == " ":
                    b[i][j] = "O"
                    best = min(best, minimax_ab(b, alpha, beta, True))
                    b[i][j] = " "
                    beta = min(beta, best)

                    if beta <= alpha:
                        break
        return best

def best_move(b):
    best_val = -math.inf
    move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if b[i][j] == " ":
                b[i][j] = "X"
                move_val = minimax_ab(b, -math.inf, math.inf, False)
                b[i][j] = " "

                if move_val > best_val:
                    best_val = move_val
                    move = (i, j)

    return move

# 🎮 GAME LOOP
print("Bạn là O | AI là X")
print("Nhập tọa độ từ 0-2 (hàng, cột)")

while True:
    print_board()

    # Người chơi
    try:
        x = int(input("Nhập hàng: "))
        y = int(input("Nhập cột: "))
    except:
        print("Nhập số hợp lệ!")
        continue

    if not (0 <= x <= 2 and 0 <= y <= 2):
        print("Ngoài phạm vi!")
        continue

    if board[x][y] != " ":
        print("Ô đã có rồi!")
        continue

    board[x][y] = "O"

    if check_winner(board):
        print_board()
        print("🎉 Bạn thắng!")
        break
    if is_full(board):
        print_board()
        print("Hòa!")
        break

    # AI đánh
    i, j = best_move(board)
    board[i][j] = "X"
    print(f"AI đánh: ({i}, {j})")

    if check_winner(board):
        print_board()
        print("🤖 AI thắng!")
        break
    if is_full(board):
        print_board()
        print("Hòa!")
        break
```
