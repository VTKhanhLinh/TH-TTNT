```python
import math

board = [[" "]*3 for _ in range(3)]

def print_board():
    for row in board:
        print("|".join(row))
        print("-"*5)

def check_winner(b):
    for row in b:
        if row[0] == row[1] == row[2] and row[0] != " ":
            return row[0]

    for col in range(3):
        if b[0][col] == b[1][col] == b[2][col] and b[0][col] != " ":
            return b[0][col]

    if b[0][0] == b[1][1] == b[2][2] and b[0][0] != " ":
        return b[0][0]
    if b[0][2] == b[1][1] == b[2][0] and b[0][2] != " ":
        return b[0][2]

    return None

def is_full(b):
    return all(cell != " " for row in b for cell in row)

def minimax(b, is_max):
    winner = check_winner(b)

    if winner == "X": return 1
    if winner == "O": return -1
    if is_full(b): return 0

    if is_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == " ":
                    b[i][j] = "X"
                    best = max(best, minimax(b, False))
                    b[i][j] = " "
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == " ":
                    b[i][j] = "O"
                    best = min(best, minimax(b, True))
                    b[i][j] = " "
        return best

def best_move(b):
    best_val = -math.inf
    move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if b[i][j] == " ":
                b[i][j] = "X"
                move_val = minimax(b, False)
                b[i][j] = " "
                if move_val > best_val:
                    best_val = move_val
                    move = (i, j)

    return move

# GAME LOOP
while True:
    print_board()

    # Người chơi (O)
    x = int(input("Nhập hàng (0-2): "))
    y = int(input("Nhập cột (0-2): "))

    if board[x][y] != " ":
        print("Ô đã chọn rồi!")
        continue

    board[x][y] = "O"

    if check_winner(board):
        print_board()
        print("Bạn thắng!")
        break
    if is_full(board):
        print_board()
        print("Hòa!")
        break

    # AI (X)
    i, j = best_move(board)
    board[i][j] = "X"

    if check_winner(board):
        print_board()
        print("AI thắng!")
        break
    if is_full(board):
        print_board()
        print("Hòa!")
        break
```
