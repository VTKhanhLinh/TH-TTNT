```python
#Bài 3. Cài đặt thuật toán minimax cho bài toán tictactoe với ma trận 5x5, 10x10, NxN.
import math

N = 5  # đổi thành 10 hoặc N bất kỳ
WIN_LEN = 4  # số quân liên tiếp để thắng (có thể chỉnh)

board = [[" "]*N for _ in range(N)]

# Kiểm tra thắng
def check_winner(b):
    # hàng
    for i in range(N):
        for j in range(N - WIN_LEN + 1):
            line = b[i][j:j+WIN_LEN]
            if line.count(line[0]) == WIN_LEN and line[0] != " ":
                return line[0]

    # cột
    for j in range(N):
        for i in range(N - WIN_LEN + 1):
            line = [b[i+k][j] for k in range(WIN_LEN)]
            if line.count(line[0]) == WIN_LEN and line[0] != " ":
                return line[0]

    # chéo \
    for i in range(N - WIN_LEN + 1):
        for j in range(N - WIN_LEN + 1):
            line = [b[i+k][j+k] for k in range(WIN_LEN)]
            if line.count(line[0]) == WIN_LEN and line[0] != " ":
                return line[0]

    # chéo /
    for i in range(WIN_LEN-1, N):
        for j in range(N - WIN_LEN + 1):
            line = [b[i-k][j+k] for k in range(WIN_LEN)]
            if line.count(line[0]) == WIN_LEN and line[0] != " ":
                return line[0]

    return None

def is_full(b):
    return all(cell != " " for row in b for cell in row)

# 🔥 Hàm đánh giá (heuristic)
def evaluate(b):
    winner = check_winner(b)
    if winner == "X":
        return 100
    elif winner == "O":
        return -100
    return 0  # có thể nâng cấp thêm

# 🔥 Minimax có depth limit
def minimax(b, depth, is_max):
    score = evaluate(b)

    if abs(score) == 100 or depth == 0 or is_full(b):
        return score

    if is_max:
        best = -math.inf
        for i in range(N):
            for j in range(N):
                if b[i][j] == " ":
                    b[i][j] = "X"
                    best = max(best, minimax(b, depth-1, False))
                    b[i][j] = " "
        return best
    else:
        best = math.inf
        for i in range(N):
            for j in range(N):
                if b[i][j] == " ":
                    b[i][j] = "O"
                    best = min(best, minimax(b, depth-1, True))
                    b[i][j] = " "
        return best

# Tìm nước đi tốt nhất
def best_move(b, depth=2):
    best_val = -math.inf
    move = (-1, -1)

    for i in range(N):
        for j in range(N):
            if b[i][j] == " ":
                b[i][j] = "X"
                move_val = minimax(b, depth, False)
                b[i][j] = " "

                if move_val > best_val:
                    best_val = move_val
                    move = (i, j)

    return move
```
